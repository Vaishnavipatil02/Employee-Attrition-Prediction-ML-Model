```python
dataset = dataset.drop([ 'Over18', 'EmployeeCount', 'EmployeeNumber', 'StandardHours'], axis = 1 )
```

# DATA PREPROCESSING

```python
#transfer data from string to integer
from sklearn.preprocessing import LabelEncoder

for column in dataset.columns:
    if dataset[column].dtype == np.number:
        continue
    dataset[column] = LabelEncoder().fit_transform(dataset[column])
```

```python
#First five rows of the encoded data frame
dataset.head()
```

# MODEL BUILDING

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

#Use Random Forest Classifire
random_forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
```

```python
X = dataset.drop(['Attrition'], axis = 1)
Y = dataset['Attrition']
```

```python
#spliting the data into training data and testing data into 80% training and 20% testing
X_train, X_test, Y_train , Y_test = train_test_split(X, Y, test_size= 0.20, random_state=0)
```

```python
X_train.head()
```

```python
#fitting data into the model to prediction
random_forest.fit(X_train, Y_train)
```

```python
random_forest.score(X_train, Y_train)
```

 Predicting test values

```python
pred = random_forest.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, pred)
```

Accuracy of the model is 85%
